```python
import pandas as pd
import numpy as np
```

```python
X_p = pd.read_csv('./인허가데이터 음식점.csv',encoding='CP949') 
```

```python
# reset 용도
X_permission = X_p.copy()
```

```python
# 여기서 필요한 정보만 추려내자
X_permission.columns
```

```python
# 5개만 보기로함
X_permission = X_permission[['인허가일자','영업상태명','폐업일자','소재지전체주소','시설총규모']]
```

```python
# 주소형식. 도로명주소가 아니다.
X_permission['소재지전체주소']
```

```python
# 대전광역시가 있을때로 한정
# 곁측치가 존재하는 이류로 na 는 제외
X_permission = X_permission[X_permission['소재지전체주소'].str.contains("대전광역시",na=False)] ; X_permission
```

```python
# 도마동 , 변동의 경우에만 고려하도록 한정
X_permission = X_permission[X_permission['소재지전체주소'].str.contains("도마동|변동",na=False)] ; X_permission
```

```python
X_permission['장사일'] = X_permission['폐업일자']
```

```python
# 폐업일자가 없는경우 '장사일' 열을 오늘로 지정하여, 오늘까지 장사중임을 표기
X_permission.loc[X_permission['폐업일자'].isna(),'장사일'] = 20200926
```

```python
X_permission['장사일'] = X_permission['장사일'].astype(str)
```

```python
# 2015년부터 데이터가 있으므로 >= 으로 2015 년 까지는 장사했던 가게들을 조사
X_permission = X_permission[ X_permission['장사일'].str[:4].astype(float) >= 2015 ] ; X_permission
```

```python
# 인덱스 초기화
X_permission.reset_index(inplace = True, drop=True) ; X_permission
```

```python
X_permission.to_csv('인허가데이터 음식점(대전).csv')
```
