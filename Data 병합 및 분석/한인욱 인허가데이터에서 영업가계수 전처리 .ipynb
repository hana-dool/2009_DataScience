```python
# 수작업으로 많은 가게들을 제외하였다.
X_permission = pd.read_csv('인허가데이터 음식점(대전)raw3.csv',encoding='CP949') ; X_permission
```

```python
X_permission.loc[:,'장사일'] = X_permission.loc[:,'폐업일자']
```

```python
# 폐업일자가 없는경우 '장사일' 열을 오늘로 지정하여, 오늘까지 장사중임을 표기
X_permission.loc[X_permission.loc[:,'폐업일자'].isna(),'장사일'] = 20200926
```

```python
X_permission.loc[:,'장사일'] = X_permission.loc[:,'장사일'].astype(int).astype(str)
```

```python
# 2015년부터 데이터가 있으므로 >= 으로 2015 년 까지는 장사했던 가게들을 조사
X_permission = X_permission[ X_permission['장사일'].str[:4].astype(float) >= 2015 ] ; X_permission
```

```python
# 인덱스 초기화
X_permission.reset_index(inplace = True, drop=True) ; X_permission
```

```python
X_per = X_permission.copy()
```

```python
X = pd.read_csv('./치킨집 현황 데이터.csv',header=None)
```

```python
X.columns = ['날짜']
```

```python
X.loc[:,'날짜2']= X.loc[:,'날짜'].str.replace("-", "")
```

```python
X.loc[:,'영업가게'] = 0
```

```python
# 인허가 일자가 2015 이하인 값들은 20151015 라고 가정한다.
for idx in range(X_per.shape[0]) :
    if X_per.loc[idx,'인허가일자'] < 20151015 :
        X_per.loc[idx,'인허가일자'] = 20151015
```

```python
# 데이터를 int 형태로 바꾸어주어야 한다.
X_per.loc[:,'장사일'] = X_per.loc[:,'장사일'].astype('int64')
```

```python
X.loc[:,'날짜2'] = X.loc[:,'날짜2'].astype('int')
```

```python
# 다음에 해당하는 기간의 날짜에 대해서는 1 을 더해준다.( 즉 그 날짜동안은 영업을 했다는 소리이므로 )
for idx in range(X_per.shape[0]) :
    X.loc[ (X_per.loc[idx,'인허가일자'] <= X.loc[:,'날짜2']) & (X.loc[:,'날짜2'] <= X_per.loc[idx,'장사일']),'영업가게' ] += 1
```

```python
# 날짜 2 는 이제 삭제해준다.
X.drop(columns=['날짜2'] , inplace = True) ; X
```

```python
X.to_csv('치킨집 운영현황.csv',encoding='CP949')
```
