```python
import pandas as pd
```

```python
import numpy as np
```

```python
X_sales = pd.read_csv('./BHC 판매 건수.csv',encoding='CP949')
```

```python
# 이렇게 sales 의 날짜 index 를 추출하여, 
# 나중에 inner merge 를 실행하므로서 치킨판매량이 있는날만 살리게 된다.
X_index = X_sales[['날짜']]
```

```python
X_temp = pd.read_csv('./대전 2015-2020 기온.csv',encoding='CP949')
```

```python
X_rain = pd.read_csv('./대전 2015-2020 강수량.csv',encoding='CP949')
```

```python
X_google = pd.read_csv('./구글트랜드(imputed).csv',encoding='CP949')
```

```python
X_naver = pd.read_csv('./네이버 데이터랩.csv',encoding='CP949')
```

```python
X_chicken = pd.read_csv('./치킨집 운영현황.csv',encoding='CP949')
```

```python
# outer 로 설정해, 일단 날짜만 일치하면 다 넣음
X = pd.merge(X_sales, X_temp, on='날짜', how='outer')
X = pd.merge(X,X_rain,on='날짜', how='outer')
X = pd.merge(X,X_google,on='날짜', how='outer')
X = pd.merge(X,X_naver,on='날짜', how='outer')
X = pd.merge(X,X_chicken,on='날짜', how='outer')
X
```

```python
X.drop(labels=['지점_x','지점_y','Unnamed: 0_x','Unnamed: 0_y'],axis=1,inplace=True)
```

```python
X = pd.merge(X, X_index, on='날짜') ; X
```

```python
X.to_csv('5조데이터(한인욱).csv',encoding='CP949')
```
