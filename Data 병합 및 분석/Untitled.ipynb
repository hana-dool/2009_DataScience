```python
import pandas as pd
import numpy as np
```

```python
Date= pd.DataFrame(pd.date_range(start='2015-11-8', end='2020-9-26') , columns=['날짜'])
```

```python
Date[['날짜']] = Date[['날짜']].astype('str')
X_index = Date[['날짜']]
```

```python
G_Trend = pd.read_csv('./구글트랜드.csv',encoding='CP949')
G_Trend
```

```python
N_Trend = pd.read_csv('./네이버 데이터랩.csv',encoding='CP949') ; N_Trend
```

```python
X = pd.merge(Date, N_Trend, on='날짜', how='outer')
```

```python
X = pd.merge(X, G_Trend , on='날짜', how='outer')
```
